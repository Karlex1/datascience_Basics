Merging and joins

```python
data1=pd.DataFrame({'key':[1,2,3,4,5,6],
        'values1':["a","b","c","d","e","f"]})
dat2=pd.DataFrame({"key":[2,3,4,5,6,7],
       "value2":["s","e","r","t","y","t"]})

print(pd.merge(data1,dat2,on="key",how='right'))#right
print(pd.merge(data1,dat2,on="key",how='left'))#left
print(pd.merge(data1,dat2,on="key",how='inner'))#inner
left_merge=pd.merge(data1,dat2,on="key",how='left',indicator=True)
left_merge_anti_join=left_merge[left_merge["_merge"]=='left_only']
print(left_merge_anti_join.drop('_merge',axis=1))
```
